數據分析處理流程：

1.去除唯一屬性

    主鍵等，對預測無幫助

2.缺失值處理

    缺失值類型：
    
    1.完全隨機缺失:數據缺失為完全隨機    
    2.隨機缺失：數據缺失為不完全隨機，和完全變量有關    
    3.完全不隨機缺失：數據缺失與不完全變量自身取值有關    
    處理缺失值的三種方法：    
    1.去除唯一屬性：有些column 屬於唯一id屬性，對於預測並沒有幫助    
    2.缺失值處理：(1)均值插補：適用於定距型的資料（資料分為定距型和非定距型）定距型的資料如：受教育成度：文盲1，小學2.....博士7，如果資料為非定距型的則使用眾數來補齊   
                (2)同類均值插補:即根據label 分類進行插補               
                (3)極大似然估計              
                （4）多重插補               
                (5)其他插值法如:隨機插值，拉格朗日差值，牛頓差值             
                （6）模型填充：如回歸、貝葉斯、隨機森林、決策樹



3.離群值處理

    離群值的處理方法（outlier）：
    第一步：先確定何為離群值？
    確認離群值方法：
    1.MAD法：絕對值差中位數法（Median Abosulte Deviation）：先求出每個因子的中位數，再得到每個因子和中位數的絕對偏差值，再得到絕對偏差值的中位數MAD,最後確定合理範圍為[中位數+-MAD]
    2.3 sigma 法：又稱為標準化法即[mean+-標準差]（通常假設檢驗數據只有隨機誤差）
    3.百分位法：即[97.5%-2.5%]
    4.其他做法如基於距離、密度、基於聚類（丟棄小簇的聚類）

4.特征編碼

    特征編碼：boolean,one-hot Encoding

5.數據標準化

    一、數據標準化的原因：
    1.如果不標準化，數量級差異較大的資料會站主導地位
    2.數量級差異會使迭代收斂變慢
    3.依賴樣本距離的算法對於數據數量級非常敏感，正規化範圍變小，更容易收斂到最優解。
    二、數據標準化的方法:
    1.min-max scaler:
        公式：new_data=(original_data-min_data)/(max_data-min_data)
    2.z-score scaler(mean 0, deviation 1):
        適用場景:當最大最小值未知時
        new_data=(original_data-mean_data)/standard devaition
    

6.特征選擇(即降維) sklearn.decomposition.

    特征選擇：

    其實此部分可以分為“變數選擇”和“特征提取”
    
    為什麼需要特征選擇？

    答：避免維度災難

    維度災難：高維空間使空間體積提高太快，導致資料稀疏。
    維度災難的解決方法：1.增加資料數量 2.資料降維
   
   


    常見的資料降維方法：
    
    過濾法（filter）：
    1.缺失值比率（Missing Values Ratio）:即如果遺漏比大於某值（如:20%）則不使用該變數
    2.低變異數篩選（Low Variance Filter）:即如果該column 的變異數小於某個閾值（即該column沒什麼代表性），就刪除
    3.高相關篩選樹（High Correlation Filter）即Pearson相關係數 ：兩個變異數之間彼此相關係數非常高時，即可則一納入分析變項


    
    包裝法（Wrapped）:
    Random-Forest,KNN,SVM
    1.隨機森林（Random Forests）:如果某屬性在每一次分類經常出現，則使用
    2.反向特征消除（Backward Feature Elimination）:即一個一個變項刪除觀察結果，刪除影響效能小者
    3.前項特征選擇（Forward Feature Elimination）:即上面反過來
    
    嵌入法（Embedded）：
    
    SVD、PCA、LDA
    1.奇異值分解（SVD）即線性變換
    2.主成分分析（PCA）即轉換為不同特征的線性組合
    3.線性判別分析（LDA）
    



*補充：

    1.one-hot encoding 造成的維度災難問題如何解決？
    one-hot encoding +主成分分析（PCA）,但如果one-hot encoding 的類別不多，不需理會。
    2.為何Tree based 的演算法不需要使用one-hot encoding 和 normalization ？
    Tree based model: Decision tree,Random forest,基於樹的Ensemble(Boosting,Bagging),Xgboost 
    因為模型不是以“距離”進行分類不需將資料化為數值，同時如果使用大量one-hot encoding 反而會使決策樹深度過深

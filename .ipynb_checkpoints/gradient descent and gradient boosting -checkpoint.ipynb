梯度下降的學習率是什麼？

    先利用鏈式法則求出損失函數的斜率（即將損失函數進行求導，得出損失函數的斜率）
    *補充：損失函數有多種類型，殘差加總總只是一種形式，但是不論如何，梯度下降都是將損失函數求導並且求出最優解
    step size=損失函數斜率(設-5.7)*learning rate（設0.1）

    則下一步大小為=斜率*learning rate
    step size= slope *learning rate
    （-5.7*0.1=-0.57）
    new intercept=0-(step size)
    0-(-0.57)=0.57
    =>下一步即為x=0.57


=>當斜率越大，下一步的大小即越大
當斜率為正，下一步向左；斜率為負，下一步向右
